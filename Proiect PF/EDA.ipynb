```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("TSLA1.csv")
```

```python
df
```

```python
df.set_index('Date', inplace = True)
df
```

```python
df['Adj Close'].plot(label = 'Tesla Adj Close price', figsize = (15,7))
#df['Open'].plot(label = 'Tesla Open price')
plt.legend()
plt.title('Tesla Stock Prices')
plt.ylabel('Stock Price')
plt.show()
```

```python
df['Volume'].plot(figsize=(15,7))
plt.title('Volume Traded by Tesla')
```

```python
df['Total Traded'] = df['Open'] * df['Volume']
df.head()
```

```python
df['Total Traded'].plot(label = 'Tesla', figsize = (15,7))
plt.legend()
plt.ylabel('Total Traded')
```

```python
from scipy.stats import jarque_bera
```

```python
M = df['Adj Close'].to_numpy()
```

```python
M
```

```python
jb_test = jarque_bera(M)
print(jb_test.statistic)
print(jb_test.pvalue)
```
